# Develop a python code to draw a line in a diagram from position (0,0) to position (8,100)

```python
import matplotlib.pyplot as plt
x = [0, 8]
y = [0, 100]
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Line from (0,0) to (8,100)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Diagram")
plt.legend()
plt.grid(True)
plt.show()
```

# Difference between pld.subplot and pld subplots in python

## Difference between plt.subplot and plt.subplots:

### #plt.subplot:
#### 1.Creates a single subplot in a figure.
#### 2.It takes three positional arguments: nrows, ncols, and index.
#### 3.Example: plt.subplot(2, 2, 1) creates a 2x2 grid of subplots and selects the first subplot.

#### #plt.subplots:
#### 1.Creates a grid of subplots and returns a figure and an array of axes objects.
#### 2.It is more flexible and allows for easier handling of multiple subplots.
#### 3.Example: fig, axes = plt.subplots(2, 2) creates a 2x2 grid and returns axes for each. 

```python
# Using plt.subplot
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)

plt.figure(figsize=(8, 6))

# Create a 2x2 grid and plot in the first subplot
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title('sin(x)')

# Second subplot
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.title('cos(x)')

# Third subplot
plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x))
plt.title('tan(x)')

# Fourth subplot
plt.subplot(2, 2, 4)
plt.plot(x, -np.sin(x))
plt.title('-sin(x)')

plt.tight_layout()
plt.show()
```

```python
# Using plt.subplots
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

# Access each subplot via the axes array
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('sin(x)')

axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title('cos(x)')

axes[1, 0].plot(x, np.tan(x))
axes[1, 0].set_title('tan(x)')

axes[1, 1].plot(x, -np.sin(x))
axes[1, 1].set_title('-sin(x)')

# Adjust layout
plt.tight_layout()
plt.show()
```

# How will you calculate the inverse and determinant of matrix. Given matrix [(7,2)(4,5)] ?
### ANS - We can use numpy to calculate the inverses and determinent of the matrix here the example of the code

```python
import numpy as np

# Define the matrix
matrix = np.array([[7, 2], [4, 5]])

# Calculate the determinant
det = np.linalg.det(matrix)

# Calculate the inverse
if det != 0:  # Ensure the determinant is non-zero
    inverse = np.linalg.inv(matrix)
else:
    inverse = "Matrix is singular, inverse does not exist."

print(f"Determinant: {det}")
print(f"Inverse: \n{inverse}")
```

# Finding the Root of x+cos(x)=0: 
### Ans- we can use scipy.optimize.root_scalar or scipy.optimize.fsolve.

```python
from scipy.optimize import fsolve
import numpy as np

# Define the function
def func(x):
    return x + np.cos(x)

# Find the root
root = fsolve(func, 0)  # Initial guess
print(f"Root: {root[0]}")
```

# For given xs and ys interpolate values from 2.1,2.2.........2.9 .
### Ans - Here we use scipy.interpolate.interp1d

```python
from scipy.interpolate import interp1d
import numpy as np

# Given xs and ys
xs = [1, 2, 3, 4, 5]
ys = [1, 4, 9, 16, 25]

# Create the interpolation function
interp_func = interp1d(xs, ys, kind='linear')

# Values to interpolate
x_new = np.arange(2.1, 3.0, 0.1)
y_new = interp_func(x_new)

print(f"Interpolated values: {y_new}")
```
